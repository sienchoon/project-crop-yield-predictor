```python
import pandas as pd

# Reference: https://www.kaggle.com/datasets/akshatgupta7/crop-yield-in-indian-states-dataset/data

raw_yield_df = pd.read_csv("crop_yield.csv")
raw_yield_df.head()
```

```python
final_yield_df = raw_yield_df.copy()

final_yield_df
```

## Categorical Feature #1: `Crop`
---

```python
count_unique_crops = final_yield_df["Crop"].value_counts()
unique_crops = final_yield_df["Crop"].unique()

print(f"Total Unique Crops in Dataset: {len(unique_crops)}")
print()
count_unique_crops
```

## Categorical Feature #2: `Season`
---

```python
unique_seasons = final_yield_df["Season"].unique()

count_unique_seasons = final_yield_df["Season"].value_counts()


print(f"Total Unique Seasons in Dataset: {len(unique_seasons)}")
print()
count_unique_seasons
```

## Categorical Feature #3: `State`
---

```python
unique_states = final_yield_df["State"].unique()

count_unique_states = final_yield_df["State"].value_counts()


print(f"Total Unique States in Dataset: {len(unique_states)}")
print()
count_unique_states
```

```python
# Drop 'State' column and save Final DataFrame to CSV
final_yield_df = final_yield_df.drop(columns = ['State'])

final_yield_df.to_csv("final_crop_yield.csv", index = False)
```
